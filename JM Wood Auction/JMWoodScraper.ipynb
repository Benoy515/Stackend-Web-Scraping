```python
from requests import get
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```python
containers = []
for i in range(1,6):
    response = get(f"https://www.jmwood.com/auctionlist.aspx?dv=2&ps=100&p={i}")
    soup = BeautifulSoup(response.text, 'html.parser')
    containers += soup.find_all('div', class_ = 'teamcard')

response = get("https://www.jmwood.com/auctionlist.aspx?dv=1")
soup = BeautifulSoup(response.text, 'html.parser')
containers += soup.find_all('div', class_ = 'teamcard')
print(len(containers))
```

```python
auctionData = {"Title": [], "Type": [], "ImageLink": [], "Location": [], "DateAndTime": [], "Timezone": [], "Status": []}
for i, auction in enumerate(containers):
    auctionData["Title"].append(auction.find("h6", class_="textcontainer-title").text)
    auctionData["Type"].append(auction.find("h6", class_="text-primary").text.replace("(", "").replace(")", ""))
    auctionData["ImageLink"].append(auction.find("img")["src"])
    auctionData["Location"].append(auction.find("div", class_="textcontainer-contact").div.text)
    auctionData["DateAndTime"].append(auction.find("strong").text)
    auctionData["Timezone"].append(auction.find("span", class_="jmwcss-auctionlist-datetime-timezone").text)
    auctionData["Status"].append("Completed" if i < 459 else "Active")

df = DataFrame(auctionData)
display(df)
df.to_csv("JMWood.csv")
```
