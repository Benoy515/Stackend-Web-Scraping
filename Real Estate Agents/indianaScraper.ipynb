```python
from requests import get, post
from bs4 import BeautifulSoup
from pandas import DataFrame, read_html

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
itemData = {"Name": [], "LicenseID": [], "Address": []}
driver = webdriver.Chrome()
driver.get("https://mylicense.in.gov/everification/")

driver.find_element(by=By.ID, value="t_web_lookup__profession_name").send_keys("Real Estate Commission")
driver.find_element(by=By.ID, value="t_web_lookup__license_type_name").send_keys("Real Estate Broker")
driver.find_element(by=By.ID, value="t_web_lookup__license_status_name").send_keys("Active")
submit_button = driver.find_element(by=By.ID, value="sch_button")
submit_button.click()
for i in range(2,651):
    print("Page", i-1)
    WebDriverWait(driver,60).until(EC.presence_of_element_located((By.ID, "datagrid_results")))

    soup = BeautifulSoup(driver.page_source, "html.parser")
    table = soup.find("table", id="datagrid_results")
    rows = table.find_all("tr")
    scrapePage(rows)
    for j in range(3):
        try:
            arr = driver.find_elements(By.TAG_NAME, "a")
            if (i-1) % 40 == 0:
                for link in arr[::-1]:
                    print(link.text)
                    if link.text == "...":
                        print("success")
                        link.click()
                        break
            else:
                for link in arr:
                    if link.text == str(i):
                        link.click()
                        break
        except:
            print(i, "Failed")
            continue
        break

driver.quit()
```

```python

def scrapePage(rows):
    for row in rows:
        if not(row.find_all("table", role="presentation")):continue
        # print(row)
        # print("-"*99)
        cells = row.find_all("td")
        # for i in cells:
        #     print(i)
        #     print()
        name, licenseID, address = cells[1].text.strip(), cells[6].text, cells[-1].text
        itemData["Address"].append(address)
        itemData["LicenseID"].append(licenseID)
        itemData["Name"].append(name)

# scrapePage(rows)
# display(DataFrame(itemData))
```
