```python
from pandas import DataFrame, read_csv, concat
ind = read_csv("csvs/indianaAgents.csv", index_col="Index")
ind = ind.rename(columns={"LicenseID": "License Number"})
ind["State"] = ["Indiana"]*len(ind)

iowa = read_csv("csvs/iowaAgents.csv", index_col="Index")
iowa["Address"] = iowa["Street"] + " " + iowa["Location"]
iowa = iowa.drop(['Street', "Location"], axis=1)
iowa = iowa.rename(columns={"ExpirationDate": "Expiration"})
iowa["State"] = ["Iowa"]*len(iowa)

nk = read_csv("csvs/nebraskaAgentsUpdated.csv", index_col="Index")
nk = nk.drop(columns=["Email"])
nk = nk.rename(columns={"Original Issue Date": "OriginalCreatedDate"})
nk["State"] = ["Nebraska"]*len(nk)

mn = read_csv("csvs/MinnesotaTest3.csv")
mn = mn.rename(columns={"City": "Address"})
mn["State"] = ["Minnesota"]*len(mn)

df = concat([ind, iowa, nk, mn])
display(df)
df.to_csv("generalAgentData.csv")
```
