```python
from requests import get
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from pandas import DataFrame, read_html, read_csv, concat

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
links = []
for i in range(1101, 1832):
    print(i)
    url = f"https://orec.us.thentiacloud.net/webs/orec/register/#/search/all/{i*10}/10/(%2522Active%2522%2520)"
    driver = webdriver.Chrome()
    driver.get(url)
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME, "table-striped")))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()

    # print(soup)
    links += [j["href"] for j in soup.find_all("a", class_="btn-single")]
# print(links)
```

```python
f = open("oklahomaLinks.txt", "r")
links = f.read().split(",")
f.close()
```

```python
itemData = {"Name": [], "License Number": [], "OriginalCreatedDate": [], "Expiration": [], "Firm Name": [], "Address": [], 
            "Phone Number": [], "License Type": [], "Status": []}
driver = webdriver.Chrome()
for count, l in enumerate(links[:]):
    url="https://orec.us.thentiacloud.net/webs/orec/register/"+ l.replace("'","").replace(" ","")
    print(count, url)
    driver.get(url)
    WebDriverWait(driver,40).until(EC.text_to_be_present_in_element((By.CLASS_NAME, "ng-binding"), " "))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    arr = read_html(driver.page_source)
    # for i in arr:display(i)

    itemData["Name"].append(I(soup.find("h2", class_="ng-binding").text).strip())
    itemData["License Number"].append(I(arr[0]["License Number"][0]))
    itemData["OriginalCreatedDate"].append(I(arr[0]["Initial Licensure Date"][0]))
    itemData["Expiration"].append(I(arr[0]["License Expiry Date"][0]))
    itemData["Firm Name"].append(I(arr[2]["Brokerage or Business Name"][0]))
    itemData["Address"].append(I(arr[2]["Street"][0])+" "+I(arr[2]["City"][0])+" "+I(arr[2]["State"][0]))
    itemData["Phone Number"].append(I(arr[2]["Telephone No."][0]))
    itemData["License Type"].append(I(arr[3]["License Category"][0]))
    itemData["Status"].append(I(arr[0]["License Status"][0]))

driver.quit()

df2 = DataFrame(itemData)
display(df2)
df2.to_csv("oklahomaAgents3.csv")
```
