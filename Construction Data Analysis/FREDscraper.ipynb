```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import io
```

```python
dataLinks = []
titles = []
for i in range(1,31):
    print(i)
    response = get(f"https://fred.stlouisfed.org/tags/series?et=&pageID={i}&t=construction%3Busa")
    soup = BeautifulSoup(response.text, 'html.parser')
    # print(soup)
    containers = soup.find_all("tr")
    # print(containers)
    for box in containers:
        # print(box.text)
        if "machinery" in box.text.lower() or "equipment" in box.text.lower() or "contract" in box.text.lower() or "spend" in box.text.lower():
            dataLinks.append("https://fred.stlouisfed.org"+box.find("a")["href"])
            titles.append(box.find("a").text)


# for i, link in enumerate(dataLinks):
#     print(i)
#     print(link)
print(len(dataLinks))
print(titles)
```

```python
num = 0
arr = []
for i, title in zip(dataLinks, titles):
    code = i.split('/')[-1]
    url = f"https://fred.stlouisfed.org/graph/fredgraph.csv?id={code}"
    print(url)
    s=get(url).content
    df=pd.read_csv(io.StringIO(s.decode('utf-8')))
    # df["Index"] = list(range(num,num+len(df)))
    df["GraphName"] = [title]*len(df)
    df.columns = ["Date", "GraphIndex", "GraphName"]
    arr.append(df)
    # df.set_index("Index", inplace=True)
    # df.set_index(list(range(num,num+len(df))),inplace=True)
    print(len(df))
    display(df)
    num+=len(df)
final = pd.concat(arr, ignore_index=True)
display(final)
final.to_csv("FRED-construction-data.csv")
```
