```python
import pandas as pd
from openpyxl import load_workbook
import os
```

```python
wb = load_workbook(filename="/Users/abhaybenoy/Main-Files/Hacking/Python_Projects/Stackend/USDA Government Payments/GP_Program.xlsx", data_only=True)
frames = []
num = 0
for name in list(wb.sheetnames)[1:]:
    print(name)
    ws = wb[name]
    data = {"Year": ["2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]}
    data["State"] = [name]*9
    for row in list(ws.rows)[5:27]:
        l = [i.value for i in row]
        l.pop(4)
        data[l[0]] = l[1:10]
        print((l))
        print("-"*50)
    df = pd.DataFrame(data, index=range(num,num+9))
    num+=9
    frames.append(df)
    # display(df)
    # df.to_csv(f"/Users/abhaybenoy/Main-Files/Hacking/Python_Projects/Stackend/USDA Government Payments/CSV_files/{name}_data.csv")
result = pd.concat(frames)
result.index.name='Index'
result.to_csv("stateData.csv")
display(result)
wb.close()
```
