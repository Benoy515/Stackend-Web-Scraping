```python
from requests import get, post
from bs4 import BeautifulSoup
from pandas import DataFrame, read_html, concat, read_csv

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url = "https://contractor.iowa.gov/IowaIWD/CREG/publicSearch/publicSearch.jsp"
soup = BeautifulSoup(get(url).text, "html.parser")

base = soup.find("select", id="principalActivity")
arr = [i.text for i in base.find_all("option")][1:]
print()
print(arr)
```

```python
finals = []
driver = webdriver.Chrome(executable_path="../chromedriver_mac_arm64/chromedriver")
for profession in arr:
    driver.get("https://contractor.iowa.gov/IowaIWD/CREG/publicSearch/publicSearch.jsp")

    driver.find_element(by=By.ID, value="principalActivity").send_keys(profession)
    submit_button = driver.find_element(by=By.ID, value="btnSearch")
    submit_button.click()

    finals.append(read_html(driver.page_source)[-1])
driver.quit()

df = concat(finals, ignore_index=True)
display(df)
```

```python
df.to_csv("iowaContractors.csv")
```
