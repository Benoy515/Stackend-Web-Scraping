```python
from requests import get, post
from bs4 import BeautifulSoup
from pandas import DataFrame, read_html, concat, read_csv

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
finals = []
for profession in ["Electricians", "Air Conditioning Contractors"]:
    driver = webdriver.Chrome(executable_path="../chromedriver_mac_arm64/chromedriver")
    driver.get("https://www.tdlr.texas.gov/LicenseSearch/")

    driver.find_element(by=By.ID, value="SelectStatus").send_keys(profession)
    submit_button = driver.find_element(by=By.ID, value="Submit1")
    submit_button.click()

    dfArr = []
    #400
    for i in range(1, 401):
        driver.get(f"https://www.tdlr.texas.gov/LicenseSearch/SearchResultsListBrowse.asp?pageNo={i}")
        dfs = read_html(driver.page_source)
        # print(driver.page_source)
        # print(dfs)
        dfArr.append(dfs[2])
    professionDF = concat(dfArr)
    professionDF["Profession"] = [profession]*len(professionDF)
    finals.append(professionDF)
    driver.close()

driver = webdriver.Chrome(executable_path="../chromedriver_mac_arm64/chromedriver")
driver.get("https://www.tdlr.texas.gov/LicenseSearch/")

driver.find_element(by=By.ID, value="SelectStatus").send_keys("Water Well Drillers,Pump Installers")
submit_button = driver.find_element(by=By.ID, value="Submit1")
submit_button.click()

dfArr = []
#400
for i in range(1, 73):
    driver.get(f"https://www.tdlr.texas.gov/LicenseSearch/SearchResultsListBrowse.asp?pageNo={i}")
    dfs = read_html(driver.page_source)
    # print(driver.page_source)
    # print(dfs)
    dfArr.append(dfs[2])
professionDF = concat(dfArr)
professionDF["Profession"] = ["Water Well Drillers/Pump Installers"]*len(professionDF)
finals.append(professionDF)
driver.close()
finalDF = concat(finals)
finalDF.to_csv("texasContractors.csv")
display(finalDF)
```
